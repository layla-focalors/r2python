다음은 조선시대 27명의 왕들에 대한 수명을 조사하여 얻은 것이다.  

태조 73, 정종 62, 태종 45, 세종 53, 문종 38, 단종 16, 세조 51, 예종 28, 성종 37, 연신군 30, 중종 56, 인종 30, 명종 33, 선조 56, 광해군 66, 인조 54, 효종 40, 현종 33, 숙종 59, 경종 36, 영조 82, 정조 48, 순조 44, 현종 22, 철종 32, 고종 67, 순종 52

1) 관찰값들을 임진왜란을 기준으로 2개의 그룹으로 나누어 변수 war를 생성하여라.
2) 임진왜란 전과 후 각각에 대하여 변수 age(나이)의 기술통계량, 바이올린 그림, 상자 그림 등을 출력하여 비교하여라.
3) 변수 war와 age의 일차원 빈도표를 각각 작성하여라.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.rc('font', family='NanumGothic')

# 데이터 생성
kings = ['태조', '정종', '태종', '세종', '문종', '단종', '세조', '예종', '성종', '연신군', '중종', '인종', '명종', '선조', '광해군', '인조', '효종', '현종', '숙종', '경종', '영조', '정조', '순조', '현종', '철종', '고종', '순종']
ages = [73, 62, 45, 53, 38, 16, 51, 28, 37, 30, 56, 30, 33, 56, 66, 54, 40, 33, 59, 36, 82, 48, 44, 22, 32, 67, 52]

df = pd.DataFrame({'왕': kings, '나이': ages})

# 1) 관찰값들을 임진왜란을 기준으로 2개의 그룹으로 나누어 변수 war를 생성하여라.
df['war'] = df['왕'].apply(lambda x: '전' if x in ['태조', '정종', '태종', '세종', '문종', '단종', '세조', '예종', '성종', '연신군', '중종', '인종', '명종', '선조'] else '후')

# 2) 임진왜란 전과 후 각각에 대하여 변수 age(나이)의 기술통계량, 바이올린 그림, 상자 그림 등을 출력하여 비교하여라.
print(df.groupby('war')['나이'].describe())

plt.figure(figsize=(12, 6))
plt.subplot(121)
sns.boxplot(x='war', y='나이', data=df)
plt.title('나이 상자 그림')
plt.subplot(122)
sns.violinplot(x='war', y='나이', data=df)
plt.title('나이 바이올린 그림')
plt.show()

# 3) 변수 war와 age의 일차원 빈도표를 각각 작성하여라.
print(df['war'].value_counts())
print(df['나이'].value_counts().sort_index())
```
