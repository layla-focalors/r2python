```python
# 06 표본평균
# 예제(6.2), p190
from scipy.stats import binom

n = 400             # 차가 지나간 수
p = 0.48            # 안전벨트를 하고 있을 확률
k1 = int(0.45 * n)  # 안전벨트를 하고 있을 비율이 45%일 확률
k2 = int(0.55 * n)  # 안전벨트를 하고 있을 비율이 55%일 확률

prob = binom.cdf(k2, n, p) - binom.cdf(k1 - 1, n, p)
print(f'안전벨트를 하고 있을 비율이 45%에서 55%일 확률 : {prob:.4f}')
```

```python
# 예제(6.2), p190 + 시각화
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

x = np.linspace(-4, 4, 1000)
y = norm.pdf(x)

fig, ax = plt.subplots()
ax.plot(x, y)

x_fill = np.linspace(-1.2, 2.8, 1000)
y_fill = norm.pdf(x_fill)
ax.fill_between(x_fill, y_fill, alpha=0.5)
plt.title("P(-1.2 <= Z <= 2.8)")
plt.axvline(-1.2, color="red", linestyle="--")
plt.axvline(2.8, color="red", linestyle="--")

plt.show()
```

```python
from statsmodels.stats.proportion import proportions_ztest

# 도시지역의 성인여성 발병률
n1 = 200 # 샘플 크기
x1 = 20 # 샘플 중 발병한 인원 수

# 농촌지역의 성인여성 발병률
n2 = 150 # 샘플 크기
x2 = 10 # 샘플 중 발병한 인원 수

# z-검정
z_stat, p_value = proportions_ztest([x1, x2], [n1, n2])
p_value = (p_value / 2)

alpha = 0.05 # 유의 수준
if p_value < alpha:
    print(f"p-value는 {p_value:.4f}로, 유의 수준 {alpha}보다 작다. \n따라서 귀무 가설을 기각하고 대립 가설을 채택한다.")
else:
    print(f"p-value는 {p_value:.4f}로, 유의 수준 {alpha}보다 크거나 같습니다. \n따라서 귀무 가설을 기각할 수 없다.")
```

```python
from scipy import stats
import numpy as np

# 귀무가설 : 피로한 상태에서 신체기능조절능력이 떨어지지 않는다.
# 대립가설 : 피로한 상태에서 신체기능조절능력이 떨어진다.

non_fatigued = np.array([158, 92, 65, 98, 33, 89, 148, 58, 142, 117, 74, 66, 109, 57, 85])
fatigued = np.array([91, 59, 215, 226, 223, 91, 92, 177, 134, 116, 153, 219, 143, 164, 100])

# t-검정
t_statistic, p_value = stats.ttest_rel(non_fatigued, fatigued)

print(f"t-statistic : {t_statistic:.4f}")
print(f"p-value: {p_value/2:.4f}\n") # 양측검정을 위해 p-value를 2로 나눔

alpha = 0.05
if p_value/2 < alpha:   
    print(f"p-value는 {p_value/2:.4f}로, 유의 수준 {alpha}보다 작다.\n따라서 귀무 가설을 기각한다.")    
else:
    print(f"p-value는 {p_value/2:.4f}로, 유의 수준 {alpha}보다 크거나 같다.\n따라서 귀무 가설을 기각할 수 없음")   
```

```python
from scipy import stats
import numpy as np

# 피로하지 않은 상태에서의 신체 기능 조절 능력
non_fatigued = np.array([158, 92, 65, 98, 33, 89, 148, 58, 142, 117, 74, 66, 109, 57, 85])

# 피로한 상태에서의 신체 기능 조절 능력
fatigued = np.array([91, 59, 215, 226, 223, 91, 92, 177, 134, 116, 153, 219, 143, 164, 100])

# 쌍체 t-검정 수행
t_statistic, p_value = stats.ttest_rel(non_fatigued, fatigued)

# t-통계량 출력
print(f"t-statistic : {t_statistic:.4f}")

# 단측 검정을 위해 p-value를 2로 나눔
p_value = (p_value / 2)
print(f"p-value: {p_value:.4f}\n")

# 유의 수준 설정
alpha = 0.05

# p-value가 유의 수준보다 작으면 귀무 가설을 기각
if p_value < alpha:   
    print(f"p-value는 {p_value:.4f}로, 유의 수준 {alpha}보다 작다.\n따라서 귀무 가설을 기각한다.")    
else:
    print(f"p-value는 {p_value:.4f}로, 유의 수준 {alpha}보다 크거나 같다.\n따라서 귀무 가설을 기각할 수 없다.")
```

```python
from scipy import stats
import numpy as np

# 귀무 가설: 피로하지 않은 상태에서의 신체 기능 조절 능력과 피로한 상태에서의 신체 기능 조절 능력은 같다.
# 대립 가설: 피로하지 않은 상태에서의 신체 기능 조절 능력이 피로한 상태에서의 신체 기능 조절 능력보다 좋다.

non_fatigued = np.array([158, 92, 65, 98, 33, 89, 148, 58, 142, 117, 74, 66, 109, 57, 85])
fatigued = np.array([91, 59, 215, 226, 223, 91, 92, 177, 134, 116, 153, 219, 143, 164, 100])

d = non_fatigued - fatigued  # 피로하지 않은 상태와 피로한 상태의 차이

# t-검정
t_stat, p_value = stats.ttest_1samp(d, popmean=0)

alpha = 0.05 # 유의 수준
if p_value / 2 < alpha:
    print(f"p-value는 {p_value / 2:.4f}로, 유의 수준 {alpha}보다 작다. \n따라서 귀무 가설을 기각하고 대립 가설을 채택한다.")
else:
    print(f"p-value는 {p_value / 2:.4f}로, 유의 수준 {alpha}보다 크거나 같다. \n따라서 귀무 가설을 기각할 수 없다.")
```

```python
from scipy import stats

# 귀무 가설: 새로운 정신안정제의 효과와 가짜약에 의한 효과는 같다.
# 대립 가설: 새로운 정신안정제의 효과가 가짜약에 의한 효과보다 좋다.

stabilizator = [19, 11, 14, 17, 23, 11, 15, 19, 11, 8]
fake = [22, 18, 17, 19, 22, 12, 14, 11, 19, 7]

d = [s - f for s, f in zip(stabilizator, fake)]  # 안정제-가짜약의 차이

# t-검정
t_stat, p_value = stats.ttest_1samp(d, popmean=0)

alpha = 0.05 # 유의 수준
if p_value / 2 < alpha:
    print(f"p-value는 {p_value / 2:.4f}로, 유의 수준 {alpha}보다 작다. \n따라서 귀무 가설을 기각하고 대립 가설을 채택한다.")
else:
    print(f"p-value는 {p_value / 2:.4f}로, 유의 수준 {alpha}보다 크거나 같다. \n따라서 귀무 가설을 기각할 수 없다.")
```
