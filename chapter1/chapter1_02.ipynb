```python
import matplotlib.pyplot as plt
import numpy as np

degree = [33, 36, 37, 38, 39, 27, 30, 35, 26, 48, 40, 41, 41, 42, 45, 35, 24, 30, 40, 42]

# 도수분포 계산
counts, bin_edges = np.histogram(degree, bins=5)

# 상대도수 계산
relative_frequency = counts / sum(counts)

# 누적상대도수 계산
cumulative_relative_frequency = np.cumsum(relative_frequency)

# 도수분포 표시
plt.hist(degree, bins=5, color='pink', alpha=0.9, edgecolor='black')
plt.title('Histogram')
plt.show()

# 상대도수 표시
plt.bar(bin_edges[:-1], relative_frequency, width = np.diff(bin_edges), color='blue', alpha=0.7, edgecolor='black')
plt.title('Relative Frequency')
plt.show()

# 누적상대도수 표시
plt.plot(bin_edges[:-1], cumulative_relative_frequency, marker='o', color='red')
plt.title('Cumulative Relative Frequency')
plt.show()

# 히스토그램
plt.hist(degree, bins=5, color='pink', alpha=0.9, edgecolor='black')
plt.title('Histogram')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

degree = [33, 36, 37, 38, 39, 27, 30, 35, 26, 48, 40, 41, 41, 42, 45, 35, 24, 30, 40, 42]

# 도수분포 계산
counts, bin_edges = np.histogram(degree, bins=5)

# 상대도수 계산
relative_frequency = counts / sum(counts)

# 누적상대도수 계산
cumulative_relative_frequency = np.cumsum(relative_frequency)

# 도수분포표 출력
print("도수분포표:")
for i in range(len(bin_edges)-1):
    print(f"{bin_edges[i]} - {bin_edges[i+1]}: {counts[i]}")

# 상대도수표 출력
print("\n상대도수표:")
for i in range(len(bin_edges)-1):
    print(f"{bin_edges[i]} - {bin_edges[i+1]}: {relative_frequency[i]}")

# 누적상대도수 출력
print("\n누적상대도수:")
for i in range(len(bin_edges)-1):
    print(f"{bin_edges[i]} - {bin_edges[i+1]}: {cumulative_relative_frequency[i]}")

# 히스토그램 출력
plt.hist(degree, bins=5, color='pink', alpha=0.9, edgecolor='black')
plt.title('Histogram')
plt.show()
```
