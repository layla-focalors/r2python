```python
import matplotlib.pyplot as plt
import numpy as np
plt.rc('font', family='Malgun Gothic')

data = [7, 10, 12, 4, 8, 7, 3, 8, 5, 12, 11, 3, 8, 1, 1, 13, 10, 4, 4, 5, 5, 8, 7, 7, 3, 2, 3, 8, 15, 1, 7, 17, 3, 4, 5, 5, 3, 1, 17, 10, 4, 7, 7, 11, 8]

# 계급의 수 결정 (통상적으로 데이터 개수의 제곱근을 사용)
num_bins = int(np.sqrt(len(data)))
hist, bin_edges = np.histogram(data, bins=num_bins)
for i in range(num_bins):
    print(f"{bin_edges[i]:.2f} - {bin_edges[i+1]:.2f}: {hist[i]}")
plt.hist(data, bins=num_bins, edgecolor='black')
plt.title('수면시간에 대한 히스토그램')
plt.xlabel('수면시간')
plt.ylabel('빈도')
plt.show()
```
