```python
import matplotlib.pyplot as plt
import numpy as np

value = [56, 89, 165, 73, 83, 145, 90, 189, 127, 77, 110, 112, 132, 120, 94, 130, 84, 65, 99, 154, 86, 120, 122, 103, 130]

# 줄기와 잎 분리
stems = [v // 10 for v in value]
leaves = [v % 10 for v in value]

# 줄기-잎 그림 생성
plt.figure(figsize=(10, 8))
(markerline, stemlines, baseline) = plt.stem(stems, leaves, linefmt='-', basefmt=' ')
plt.setp(markerline, 'markerfacecolor', 'b')
plt.setp(baseline, 'color', 'r', 'linewidth', 2)

plt.show()
```
